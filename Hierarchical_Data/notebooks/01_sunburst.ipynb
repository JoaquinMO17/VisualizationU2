```python
import pandas as pd
import pycountry_convert as pc
df = pd.read_csv('Demographic.csv')

df = df.iloc[:435]
# Reemplazar '..' con NaN para que pandas pueda convertir a número
df = df.replace('..', pd.NA)
df = df.dropna(subset=['2024 [YR2024]'])
df['2024 [YR2024]'] = df['2024 [YR2024]'].astype(float)

# Pivotear el dataset: queremos tener columnas "Population" y "GDP per capita"
df_pivot = df.pivot_table(
    index=["Country Name", "Country Code"],
    columns="Series Name",
    values="2024 [YR2024]"
).reset_index()

# Renombrar columnas a algo más manejable
df_pivot.columns.name = None
df_pivot = df_pivot.rename(columns={
    "Population, total": "Population",
    "GDP per capita (current US$)": "GDP_per_capita"
})

# Eliminar filas con datos faltantes
df_pivot = df_pivot.dropna(subset=["Population", "GDP_per_capita"])

latin_america_iso3 = {
    "ARG", "BOL", "BRA", "CHL", "COL", "CRI", "CUB", "DOM", "ECU", "SLV",
    "GTM", "HND", "MEX", "NIC", "PAN", "PRY", "PER", "URY", "VEN"
}
def get_region_from_iso3(iso3):
    try:
        country_alpha2 = pc.country_alpha3_to_country_alpha2(iso3)
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        # Clasificar América Latina aparte
        if iso3 in latin_america_iso3:
            return "Latin America"
        elif continent_name == "North America":
            return "North America (Non-Latin)"
        elif continent_name == "South America":
            return "South America (Non-Latin)"
        elif continent_name == "Asia":
            return "Asia"
        else:
            return continent_name
    except:
        return "Other"

df_pivot["Region"] = df_pivot["Country Code"].apply(get_region_from_iso3)

df_focus = df_pivot[df_pivot["Region"].isin(["Latin America", "Asia"])]

df_focus.head()
```

```python
import plotly.express as px

fig = px.sunburst(
    df_focus,
    path=["Region", "Country Name"],
    values="Population",
    color="GDP_per_capita",
    color_continuous_scale="Viridis",
    hover_data={
        "Population": ":,.0f",
        "GDP_per_capita": ":,.0f"
    },
    title="Evaluación de Mercados: Asia y América Latina (2024)"
)

fig.update_layout(
    margin=dict(t=50, l=0, r=0, b=0),
    coloraxis_colorbar=dict(title="PIB per cápita (USD)")
)

fig.show()
```
